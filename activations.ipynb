```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load the activations from the .npy file
activations = np.load("data/activations.npy", allow_pickle=True)

# Shape of the first convolutional layer activations
print(activations[0].shape)

# Plotting the 1D heatmap
plt.figure(figsize=(12, 6))

# Create the heatmap from the first convolutional layer activations
sns.heatmap(
    activations[0],
    cmap="Greys",
    cbar=True,
    xticklabels=50,
    yticklabels=[f"Kernel {i+1}" if i % 2 == 0 else "" for i in range(activations[0].shape[0])], 
    square=False,
    linecolor='black',
)

# Set plot labels and title
plt.xlabel("Feature Index")
plt.ylabel("Output Channels (Kernels)")
plt.title("1D Heatmap of First Convolutional Layer Activations")
plt.show()
```

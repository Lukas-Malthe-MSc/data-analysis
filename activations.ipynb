```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the real scan data from the .npy file
real_scan_data = np.load('data/scan_frame.npy', allow_pickle=True).item()
real_ranges = real_scan_data['ranges']
real_range_min = real_scan_data['range_min']
real_range_max = real_scan_data['range_max']

# Clip and normalize the real scan ranges
real_ranges_clipped = np.clip(real_ranges, real_range_min, real_range_max)
real_ranges_normalized = (real_ranges_clipped - real_range_min) / (real_range_max - real_range_min)

# Load the simulated scan data from the .npy file
sim_ranges_normalized = np.load("data/lidar.npy", allow_pickle=True)

# Create a row of NaN values to act as a separator
separator = np.full(real_ranges_normalized.shape, np.nan)

# Combine the data for plotting
combined_data = np.vstack((real_ranges_normalized, separator, sim_ranges_normalized))

# Create the heat map
plt.figure(figsize=(10, 3))
ax = sns.heatmap(
    combined_data,
    cmap="RdYlGn",
    cbar=True,
    xticklabels=200,
    yticklabels=['Real Scan', '', 'Simulated Scan'],
    square=False
)
ax.collections[0].colorbar.set_label('Normalized Range')
plt.title('1D Heat Map of Real and Simulated Laser Scan Ranges')
plt.xlabel('Beam Index')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
lidar = np.load("data/lidar.npy", allow_pickle=True)
activations = np.load("data/activations.npy", allow_pickle=True)

# Function to extract subsets of activations
def extract_subset(activations, layer_idx, mod_value):
    return activations[layer_idx][::mod_value]

# Configuration variables
cmap = "RdYlGn"
figsize = (8, 6)
height_ratios = [0.3, 1, 1, 1]
xticklabels = [100, 50, 25, 10]
ytick_mod_values = [2, 4, 4]
layer_titles = [
    "LIDAR Distances Heatmap (1D)",
    "1st Convolutional Layer",
    "2nd Convolutional Layer",
    "3rd Convolutional Layer"
]

# Extract subsets of the activations for specific kernels
subset_kernels_mod4 = extract_subset(activations, 0, 2)  # Kernels where index % 4 == 0
subset_kernels_mod8 = extract_subset(activations, 3, 4)  # Kernels where index % 8 == 0
subset_kernels_mod8_layer3 = extract_subset(activations, 6, 4)  # Kernels where index % 8 == 0 for the third layer

# Combine all data to find the overall min and max for the color scale
all_data = np.concatenate([lidar.flatten(), subset_kernels_mod4.flatten(), subset_kernels_mod8.flatten(), subset_kernels_mod8_layer3.flatten()])
abs_max = max(abs(all_data.min()), abs(all_data.max()))
vmin = -abs_max
vmax = abs_max

# Create a figure with multiple subplots
fig, axes = plt.subplots(4, 1, figsize=figsize, gridspec_kw={'height_ratios': height_ratios})

# Plot the LIDAR heatmap
sns.heatmap(lidar.reshape(1, -1), 
            cmap=cmap,
            vmin=vmin,
            vmax=vmax,
            cbar=False,
            xticklabels=xticklabels[0],
            yticklabels=False,
            ax=axes[0]
)
axes[0].set_title(layer_titles[0])

# Function to plot each convolutional layer's heatmap
def plot_layer_heatmap(ax, data, title, xticks, yticks, ytick_mod_value):
    sns.heatmap(data,
                cmap=cmap,
                vmin=vmin,
                vmax=vmax,
                cbar=False,
                xticklabels=xticks,
                yticklabels=[f"{i*ytick_mod_value}" if i%3 == 0 else '' for i in range(data.shape[0])],
                square=False,
                ax=ax)
    ax.set_title(title)

# Plot the subset of kernels (mod 4) from the first convolutional layer
plot_layer_heatmap(axes[1], subset_kernels_mod4, layer_titles[1], xticklabels[1], True, ytick_mod_values[0])

# Plot the subset of kernels (mod 8) from the second convolutional layer
plot_layer_heatmap(axes[2], subset_kernels_mod8, layer_titles[2], xticklabels[2], True, ytick_mod_values[1])
axes[2].set_ylabel("Output Channels (Kernels)")

# Plot the subset of kernels (mod 8) from the third convolutional layer
plot_layer_heatmap(axes[3], subset_kernels_mod8_layer3, layer_titles[3], xticklabels[3], True, ytick_mod_values[2])
axes[3].set_xlabel("Feature Index")

# Invert x-axis for all subplots
for ax in axes:
    ax.invert_xaxis()

# Create a single colorbar for all subplots
cbar_ax = fig.add_axes([0.92, 0.3, 0.02, 0.4])  # Position: [left, bottom, width, height]
norm = plt.Normalize(vmin=vmin, vmax=vmax)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
fig.colorbar(sm, cax=cbar_ax)

plt.tight_layout(rect=[0, 0, 0.9, 1])
plt.subplots_adjust(hspace=1)  # Adjust spacing to fit everything neatly
plt.savefig('images/activations.svg', format='svg')
plt.show()
```

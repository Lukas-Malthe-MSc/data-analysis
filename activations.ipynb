```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the LIDAR data
lidar = np.load("data/lidar.npy", allow_pickle=True)

# Load the activations
activations = np.load("data/activations.npy", allow_pickle=True)

# Extract subsets of the activations for specific kernels
subset_kernels_mod4 = activations[0][::2]  # Kernels where index % 4 == 0
subset_kernels_mod8 = activations[3][::4]  # Kernels where index % 8 == 0
subset_kernels_mod8_layer3 = activations[6][::4]  # Kernels where index % 8 == 0 for the third layer

# Create a figure with multiple subplots
fig, axes = plt.subplots(4, 1, figsize=(8, 6), gridspec_kw={'height_ratios': [0.3, 1, 1, 1]})

# Plot the LIDAR heatmap
sns.heatmap(lidar, 
            cmap="RdYlGn",
            cbar=False,
            xticklabels=100,
            yticklabels=False,
            ax=axes[0]
)
axes[0].set_title("LIDAR Distances Heatmap (1D)")

# Plot the subset of kernels (mod 4) from the first convolutional layer
sns.heatmap(
    subset_kernels_mod4,
    cmap="RdYlGn",
    cbar=False,
    xticklabels=50,
    yticklabels=[f"{i+1}" if (i % 2 == 0) else '' for i in range(subset_kernels_mod4.shape[0])],
    square=False,
    ax=axes[1]
)
# axes[1].set_xlabel("Feature Index")
# axes[1].set_ylabel("Output Channels (Kernels)")
axes[1].set_title("1st Convolutional Layer")

# Plot the subset of kernels (mod 8) from the second convolutional layer
sns.heatmap(
    subset_kernels_mod8,
    cmap="RdYlGn",
    cbar=False,
    xticklabels=25,
    yticklabels=[f"{i+1}" if (i % 4 == 0) else '' for i in range(subset_kernels_mod8.shape[0])],
    square=False,
    ax=axes[2]
)
# axes[2].set_xlabel("Feature Index")
axes[2].set_ylabel("Output Channels (Kernels)")
axes[2].set_title("2nd Convolutional Layer")

# Plot the subset of kernels (mod 8) from the third convolutional layer
sns.heatmap(
    subset_kernels_mod8_layer3,
    cmap="RdYlGn",
    cbar=False,
    xticklabels=10,
    yticklabels=[f"{i+1}" if (i % 4 == 0) else '' for i in range(subset_kernels_mod8_layer3.shape[0])],
    square=False,
    ax=axes[3]
)
axes[3].set_xlabel("Feature Index")
# axes[3].set_ylabel("Output Channels (Kernels)")
axes[3].set_title("3rd Convolutional Layer")

plt.tight_layout()
plt.subplots_adjust(hspace=1)  # Adjust spacing to fit everything neatly

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the real scan data from the .npy file
real_scan_data = np.load('data/scan_frame.npy', allow_pickle=True).item()
real_ranges = real_scan_data['ranges']
real_range_min = real_scan_data['range_min']
real_range_max = real_scan_data['range_max']

# Clip and normalize the real scan ranges
real_ranges_clipped = np.clip(real_ranges, real_range_min, real_range_max)
real_ranges_normalized = (real_ranges_clipped - real_range_min) / (real_range_max - real_range_min)

# Load the simulated scan data from the .npy file
sim_ranges_normalized = np.load("data/lidar.npy", allow_pickle=True)

# Create a row of NaN values to act as a separator
separator = np.full(real_ranges_normalized.shape, np.nan)

# Combine the data for plotting
combined_data = np.vstack((real_ranges_normalized, separator, sim_ranges_normalized))

# Create the heat map
plt.figure(figsize=(10, 3))
ax = sns.heatmap(
    combined_data,
    cmap="RdYlGn",
    cbar=True,
    xticklabels=200,
    yticklabels=['Real Scan', '', 'Simulated Scan'],
    square=False
)
ax.collections[0].colorbar.set_label('Normalized Range')
plt.title('1D Heat Map of Real and Simulated Laser Scan Ranges')
plt.xlabel('Beam Index')
plt.show()
```

```python
# plot the rewards
import numpy as np
import matplotlib.pyplot as plt

data_files = [
    'data/test_track_1_MLP_rewards.npy',
    'data/test_track_1_LSTM_rewards.npy',
    'data/test_track_1_CNN_rewards.npy',
    'data/test_track_2_MLP_rewards.npy',
    'data/test_track_2_LSTM_rewards.npy',
    'data/test_track_2_CNN_rewards.npy',
    'data/test_track_3_MLP_rewards.npy',
    'data/test_track_3_LSTM_rewards.npy',
    'data/test_track_3_CNN_rewards.npy',
]

dt = 0.025

for data_file in data_files:
    rewards = np.load(data_file)
    rewards_accum = np.cumsum(rewards)
    # plt.plot(rewards, label=data_file)
    print(rewards_accum[-1])
    plt.plot(rewards_accum, label=data_file + ' accum')
plt.legend()
plt.show()

```

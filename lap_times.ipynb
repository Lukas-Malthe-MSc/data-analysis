```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Rewards data
rewards = {
    'Test Track 1': [1014.747, 1025.714, 1012.967],
    'Test Track 2': [1023.404, 970.626, 1033.112],
    'Test Track 3': [1114.281, 1108.968, 1089.495]
}

# Average lap times data
lap_times = {
    'Test Track 1': [13.92, 13.66, 14.15],
    'Test Track 2': [10.07, 10.58, 9.97],
    'Test Track 3': [10.04, 9.97, 10.40]
}

models = ['MLP', 'LSTM', 'CNN']
colors = ['#e41a1c', '#4daf4a', '#377eb8']  # Colorblind-friendly palette

# Calculate average rewards and lap times for each model
avg_rewards = []
avg_lap_times = []

for model_index in range(len(models)):
    model_rewards = [rewards[track][model_index] for track in rewards]
    model_lap_times = [lap_times[track][model_index] for track in lap_times]
    avg_rewards.append(np.mean(model_rewards))
    avg_lap_times.append(np.mean(model_lap_times))

# Set up the plot
plt.figure(figsize=(10, 6))

# Scatter plot
for i, model in enumerate(models):
    plt.scatter(avg_rewards[i], avg_lap_times[i], color=colors[i], label=model, s=100)

# Linear regression
slope, intercept, r_value, p_value, std_err = linregress(avg_rewards, avg_lap_times)
line = slope * np.array(avg_rewards) + intercept
plt.plot(avg_rewards, line, color='gray', linestyle='--', linewidth=2)

# Display correlation coefficient and p-value
plt.text(0.05, 0.95, f'R={r_value:.2f}\nP={p_value:.2e}', transform=plt.gca().transAxes, fontsize=12,
         verticalalignment='top', bbox=dict(boxstyle='round,pad=0.5', alpha=0.1))

# Set labels and title
plt.xlabel('Average Rewards')
plt.ylabel('Average Lap Time (s)')
plt.title('Correlation Between Average Rewards and Average Lap Times Across Models')
plt.legend(loc='upper right')

# Adjust layout and save the plot
plt.tight_layout()
plt.savefig('images/average_rewards_vs_lap_times.svg', format='svg')
plt.show()

# Print average rewards and lap times
print("Average Rewards and Lap Times:")
for i, model in enumerate(models):
    print(f"{model} - Average Reward: {avg_rewards[i]:.2f}, Average Lap Time: {avg_lap_times[i]:.2f}")

```

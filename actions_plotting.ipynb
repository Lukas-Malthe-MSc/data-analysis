```python
import numpy as np

# Load the numpy file 
file_path = 'data/actions_no_pen.npy'
actions = np.load(file_path)

# Display the data type, shape, and first few elements
print("Data Type:", type(actions))
print("Data Shape:", actions.shape)
print("First few elements:", actions[:5]) 
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Load the numpy file
file_path = 'data/actions_no_pen.npy'
actions = np.load(file_path)

# Extract the two components
speed_no_pen = actions[:, 0, 0]
sa_no_pen = actions[:, 0, 1]

# Create a time array
time = np.arange(len(actions))

# Plotting the components over time
plt.figure(figsize=(12, 6))

# Plot component 1
plt.subplot(2, 1, 1)  # 2 rows, 1 column, 1st subplot
plt.plot(time, speed_no_pen, 'b-', label='Component 1')  # 'b-' is a blue line
plt.xlabel('Time')
plt.ylabel('Component 1')
plt.title('Component 1 over Time')
plt.grid(True)
plt.legend()

# Plot component 2
plt.subplot(2, 1, 2)  # 2nd subplot
plt.plot(time, sa_no_pen, 'r-', label='Component 2')  # 'r-' is a red line
plt.xlabel('Time')
plt.ylabel('Component 2')
plt.title('Component 2 over Time')
plt.grid(True)
plt.legend()

# Display the plots
plt.tight_layout()  # Improves spacing between plots
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec

# Set the Seaborn style
plt.style.use("seaborn-v0_8-ticks")
plt.rcParams['mathtext.fontset'] = 'stix'  # Set the font for math text
plt.rcParams['font.size'] = 14  # Increase the global font size

# Load the numpy file for 'no penalty' - Actions
actions_file_path = 'data/actions_no_pen.npy'
actions = np.load(actions_file_path)
sa_no_pen = actions[:, 0, 1]  # Assuming you are extracting correct indices

# Load the numpy file for 'log penalty' - Actions
actions_log_file_path = 'data/actions_log_pen.npy'
actions_log = np.load(actions_log_file_path)
sa_log_pen = actions_log[:, 0, 1]  # Assuming you are extracting correct indices

# Load the numpy file for 'no penalty' - Position
pos_file_path = 'data/pos_data_no_pen.npy'
pos_data = np.load(pos_file_path)
velocity_no_pen = pos_data[:,1, 0, 0]  # Assuming you are extracting correct indices

# Load the numpy file for 'log penalty' - Position
pos_log_file_path = 'data/pos_data_log_pen.npy'
pos_log_data = np.load(pos_log_file_path)
velocity_log_pen = pos_log_data[:,1, 0, 0]  # Assuming you are extracting correct indices

# Time step duration in seconds
time_step_duration = 0.025  # 0.025s per time step
# Create a time array scaled by the time step duration
time = np.arange(len(actions)) * time_step_duration

# Creating a single figure with customized subplot sizes
plt.figure(figsize=(12, 6))
gs = gridspec.GridSpec(2, 1, height_ratios=[3, 5])  # Height ratio 1:2 for the two subplots

# Plot component 1: Actions over Time
ax1 = plt.subplot(gs[0])
ax1.plot(time, sa_no_pen, label=r'$\delta_d, \mathrm{no\, penalty}$', linewidth=2)
ax1.plot(time, sa_log_pen, label=r'$\delta_d, \mathrm{log\, penalty}$', linewidth=2)
ax1.set_ylabel(r'$\delta_d (\mathrm{rad})$')
ax1.set_yticks([-np.pi/4, 0, np.pi/4], [r'-$\frac{\pi}{4}$', '0', r'$\frac{\pi}{4}$'])
ax1.grid(axis='y')
ax1.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Plot component 2: Velocity over Time
ax2 = plt.subplot(gs[1])
ax2.plot(time, velocity_no_pen, label=r'$\dot{x}, \mathrm{no\, penalty}$', linewidth=2)
ax2.plot(time, velocity_log_pen, label=r'$\dot{x}, \mathrm{log\, penalty}$', linewidth=2)
ax2.plot(time, np.ones(time.shape)*5, linestyle='--', color='black', label=r'$\mathrm{Max\, speed}: 5\, \frac{m}{s}$', linewidth=1)  # Black dotted line at y=5
ax2.grid(axis='y')
ax2.set_xlabel(r'$\mathrm{Time} (s)$')
ax2.set_ylabel(r'$\dot{x}$ $\left(\frac{m}{s}\right)$')
ax2.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Display the plots
plt.tight_layout()  # Adjust layout to prevent overlap
plt.savefig('images/no_pen_vs_log_pen.svg', format='svg')
plt.show()
```

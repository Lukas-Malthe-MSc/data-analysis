```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
from matplotlib.transforms import Affine2D
from scipy.ndimage import rotate
from matplotlib.legend_handler import HandlerPatch

plt.style.use('seaborn')
plt.rcParams.update({
    'font.size': 24,
    'axes.titlesize': 16,     # Font size for titles
    'axes.labelsize': 16,     # Font size for x and y labels
    'xtick.labelsize': 12,    # Font size for x tick labels
    'ytick.labelsize': 12,    # Font size for y tick labels
    'legend.fontsize': 12     # Font size for legend
})
# List of data files to load
data_files = ['data/pos_data_cnn_no_dr.npy',
              'data/pos_data_cnn.npy']

# Define colors and labels for the plots
colors = ['#984ea3', '#377eb8']  # Adding a purple color
labels = ['CNN without Domain Randomization', 'CNN with Domain Randomization']

image_for_overlay = 'images/test_track_2_overlay.png'

plt.figure(figsize=(8, 6))  # Adjust figure size

for file_path, color, label in zip(data_files, colors, labels):
    # Load the numpy file
    pos_data = np.load(file_path)
    
    # Extract the x and y coordinates
    x_values = pos_data[1:, 0, 0, 0]
    y_values = pos_data[1:, 0, 0, 1]
    
    # Plot the (x, y) data
    plt.plot(x_values[255:], y_values[255:], color, alpha=0.8, label=label)

# Add the image overlay
img = mpimg.imread(image_for_overlay)
img = rotate(img, -90)
img_height, img_width = img.shape[:2]

# Calculate the aspect ratio
aspect_ratio = img_width / img_height

# Define the extent of the image overlay based on the aspect ratio and scale
scale_x = 8.15
scale_y = scale_x / aspect_ratio 
image_extent = [-scale_x, scale_x, -scale_y, scale_y]

print(image_extent)
plt.imshow(img, extent=image_extent, alpha=1, zorder=0)

# Add labels and a title
plt.xlabel(r'$x \, (m)$', fontsize=12)
plt.ylabel(r'$y \, (m)$', fontsize=12)
# plt.title('Line Plot of Position Data (x, y)', fontsize=14)

# Add gridlines for better readability
plt.grid(True, linestyle='--', linewidth=0.5)

# Draw a starting zone at (5,5)
# Draw a circle with a radius of 0.2 around (5,5)
start_x, start_y = 5, 5
radius = 0.2
circle = plt.Circle((start_x, start_y), radius, color='r', fill=True, linestyle='-', linewidth=2, label='Starting Zone')
plt.gca().add_patch(circle)

# Custom legend handler for the circle
class HandlerCircle(HandlerPatch):
    def create_artists(self, legend, orig_handle, xdescent, ydescent, width, height, fontsize, trans):
        circle = plt.Circle((0.5 * width - xdescent, 0.5 * height - ydescent), 0.2 * width, 
                            facecolor=orig_handle.get_facecolor(), edgecolor=orig_handle.get_edgecolor(),
                            lw=orig_handle.get_linewidth(), linestyle=orig_handle.get_linestyle(), transform=trans)
        return [circle]

# Add a legend below the plot
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=1, handler_map={circle: HandlerCircle()}, fontsize=12)

# Adjust layout and save the figure
plt.tight_layout()
plt.savefig('images/no_dr_vs_dr_trajectory.svg', format='svg')
# plt.savefig('images/no_dr_vs_dr_trajectory.pdf', format='pdf')

# Display the plot
plt.show()
```

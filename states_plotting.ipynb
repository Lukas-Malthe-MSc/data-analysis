```python
import matplotlib.pyplot as plt
import numpy as np

# Load the numpy file
file_path = 'data/pos_data_no_pen.npy'
pos_data = np.load(file_path)

# Extract the x and y coordinates
x_values = pos_data[150:250, 0, 0]
y_values = pos_data[150:250, 0, 1]

# Create a plot for the (x, y) data
plt.figure(figsize=(10, 6))  # Adjust figure size
plt.plot(x_values, y_values, 'b-', alpha=0.5, label='Position Data')  # 'b-' for a blue line with transparency

# Add labels and a title
plt.xlabel('X Coordinate', fontsize=12)
plt.ylabel('Y Coordinate', fontsize=12)
plt.title('Line Plot of Position Data (x, y)', fontsize=14)

# Add gridlines for better readability
plt.grid(True, linestyle='--', linewidth=0.5)

# Add a legend
plt.legend()

# Draw a starting zone at (5,5)
# Draw a circle with a radius of 0.2 around (5,5)
start_x, start_y = 5, 5
radius = 0.2
circle = plt.Circle((start_x, start_y), radius, color='r', fill=False, linestyle='--', linewidth=2, label='Starting Zone')
plt.gca().add_patch(circle)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg  # For loading images

# Load the numpy file
file_path = 'data/pos_data_log_pen.npy'
pos_data = np.load(file_path)

# Extract the x and y coordinates
x_values = pos_data[1:, 0, 0]
y_values = pos_data[1:, 0, 1]

# Load the map image
map_image = mpimg.imread('track_map.png')  # Replace with your map image

# Create a plot for the map and position data
plt.figure(figsize=(10, 6))

# Plot the map
xmin, xmax = -7, 7  # Adjust based on the map image
ymin, ymax = -5, 5  # Adjust based on the map image
plt.imshow(map_image, extent=[xmin, xmax, ymin, ymax], aspect='auto', alpha=1)  # Adjust extent to fit the map

# Plot the position data over the map
plt.plot(x_values, y_values, 'b-', alpha=1.0, label='Position Data')

# Add labels and a title
plt.xlabel('X Coordinate', fontsize=12)
plt.ylabel('Y Coordinate', fontsize=12)
plt.title('Line Plot of Position Data (x, y) with Map Overlay', fontsize=14)

# Add gridlines and legend
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()

# Draw a starting zone at (5,5)
start_x, start_y = 5, 5
radius = 0.2
circle = plt.Circle((start_x, start_y), radius, color='r', fill=False, linestyle='--', linewidth=2, label='Starting Zone')
plt.gca().add_patch(circle)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec
# Set the Seaborn style
plt.style.use("seaborn")
# plt.rcParams['mathtext.fontset'] = 'stix'  # Set the font for math text
plt.rcParams.update({
    'font.size': 14,         # Default font size
    'axes.labelsize': 16,    # Font size for x and y labels
    'xtick.labelsize': 12,   # Font size for x tick labels
    'ytick.labelsize': 12,   # Font size for y tick labels
    'legend.fontsize': 14    # Font size for legend
})
```

```python
# Define the range of x_dot values
x_dot = np.linspace(-5, 5, 400)

# Calculate R_velocity using the max function
R_velocity = np.maximum(x_dot, 0)

# Create the plot
plt.figure(figsize=(6, 3))
plt.plot(x_dot, R_velocity, label=r'$R_{\mathrm{velocity}} = \max(\dot{x}, 0)$', color='red', linewidth=2)
plt.xlabel(r'$\dot{x}_t$', fontsize=16)
plt.ylabel(r'$r^{\mathrm{velocity}}_t$', fontsize=16)
plt.grid(axis='y')
plt.savefig('images/R_velocity_weighted.svg', format='svg')
plt.show()
```

```python

# Define the range of delta_d values
delta_d = np.linspace(-np.pi/8, np.pi/8, 400)

# Calculate R_delta_d
R_delta_d = np.log(np.abs(180/np.pi * delta_d) + 1) * -2/8
# R_delta_d = ((delta_d)**2)
R_delta_d_prior = np.log(abs(delta_d)+1)

# Create the plot
plt.figure(figsize=(6, 3))
plt.plot(delta_d, R_delta_d, label=r'$R_{\delta_d} = \log(|\delta_d| + 1)$', color='red', linewidth=2)
# plt.plot(delta_d, R_delta_d_prior, label=r'$R_{\delta_d} = \log(|\delta_d| + 1)$', color='red', linewidth=2)

# Setting custom x-axis ticks and labels
ticks = [-np.pi/8, 0, np.pi/8]
labels = [r'$-\frac{\pi}{8}$', r'$0$', r'$\frac{\pi}{8}$']
plt.xticks(ticks, labels, fontsize=14)

plt.xlabel(r'$\delta_t$', fontsize=16)
plt.ylabel(r'$r^{\delta}_t$', fontsize=16)
plt.grid(axis='y')
plt.savefig('images/R_delta_weighted.svg', format='svg')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Collision detection results array
R_collision = [0, -10, -10]  # Ensure the step at the middle

# Corresponding states on the x-axis extended to 2
states = [0, 1, 2]

# Create the plot
plt.figure(figsize=(6, 3))
plt.step(states, R_collision, where='post', label='Step Transition at 1', color='red', linewidth=2)
plt.xticks([0, 1, 2], ['0', '1', ''], fontsize=14)  # Label the middle point as 'True'
plt.yticks([0, -10], ['0', '-10'], fontsize=14)
# plt.xlabel('Index')
plt.ylabel(r'$r^{\mathrm{collision}}_t$', fontsize=16)

plt.grid(axis='y')
plt.savefig('images/R_collision_weighted.svg', format='svg')
plt.show()
```
